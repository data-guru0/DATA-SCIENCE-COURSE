### STANDARDIZATION

```python
import numpy as np
import pandas as pd
```

```python
ages = np.random.normal(loc=40, scale=10, size=1000)
salaries = np.random.normal(loc=50000, scale=15000, size=1000)
```

```python
data_df = pd.DataFrame({'Age': ages, 'Salary': salaries})
```

```python
data_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data_df["Salary"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

standardized_data = scaler.fit_transform(data_df)

standardized_df = pd.DataFrame(standardized_data, columns=['Age', 'Salary'])
```

```python
standardized_data
```

```python
data_df.head()
```

```python
standardized_df.head()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(5,5))

# Before standardization
ax[0, 0].hist(data_df['Age'], bins=30, color='b', alpha=0.7)
ax[0, 0].set_title('Age Before Standardization')

ax[0, 1].hist(data_df['Salary'], bins=30, color='r', alpha=0.7)
ax[0, 1].set_title('Salary Before Standardization')

# After standardization
ax[1, 0].hist(standardized_df['Age'], bins=30, color='b', alpha=0.7)
ax[1, 0].set_title('Age After Standardization')

ax[1, 1].hist(standardized_df['Salary'], bins=30, color='r', alpha=0.7)
ax[1, 1].set_title('Salary After Standardization')

plt.tight_layout()
plt.show()
```

### NORMALIZATION

```python
data = np.random.uniform(low=10, high=100, size=(1000, 2))
data_df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])
```

```python
data_df.head()
```

```python
plt.hist(data_df["Feature2"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(data_df)
                                  
scaled_df = pd.DataFrame(scaled_data, columns=['Feature1', 'Feature2'])
```

```python
scaled_data
```

```python
data_df.head(3)
```

```python
scaled_df.head(3)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(5,5))

# Before scaling
ax[0, 0].hist(data_df['Feature1'], bins=30, color='b', alpha=0.7)
ax[0, 0].set_title('Feature1 Before Scaling')

ax[0, 1].hist(data_df['Feature2'], bins=30, color='r', alpha=0.7)
ax[0, 1].set_title('Feature2 Before Scaling')

# After scaling
ax[1, 0].hist(scaled_df['Feature1'], bins=30, color='b', alpha=0.7)
ax[1, 0].set_title('Feature1 After Scaling')

ax[1, 1].hist(scaled_df['Feature2'], bins=30, color='r', alpha=0.7)
ax[1, 1].set_title('Feature2 After Scaling')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Set random seed for reproducibility
np.random.seed(0)

# Generate data with outliers
n_samples = 1000

# Feature 1: Normally distributed
feature1 = np.random.normal(loc=50, scale=10, size=n_samples)

# Feature 2: Uniformly distributed with outliers
feature2 = np.random.uniform(low=10, high=100, size=n_samples)
feature2[0] = 200  # Introduce an outlier

# Combine features into a DataFrame
data = pd.DataFrame({'Feature1': feature1, 'Feature2': feature2})

# Plotting the original data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(data['Feature1'], data['Feature2'])
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# StandardScaler
scaler_standard = StandardScaler()
data_standard_scaled = scaler_standard.fit_transform(data)
data_standard_scaled = pd.DataFrame(data_standard_scaled, columns=['Feature1', 'Feature2'])

# MinMaxScaler
scaler_minmax = MinMaxScaler()
data_minmax_scaled = scaler_minmax.fit_transform(data)
data_minmax_scaled = pd.DataFrame(data_minmax_scaled, columns=['Feature1', 'Feature2'])

# Plotting after scaling
plt.subplot(1, 2, 2)
plt.scatter(data_standard_scaled['Feature1'], data_standard_scaled['Feature2'], label='StandardScaler', alpha=0.5)
plt.scatter(data_minmax_scaled['Feature1'], data_minmax_scaled['Feature2'], label='MinMaxScaler', alpha=0.5)
plt.title('Data After Scaling')
plt.xlabel('Scaled Feature 1')
plt.ylabel('Scaled Feature 2')
plt.legend()

plt.tight_layout()
plt.show()
```
